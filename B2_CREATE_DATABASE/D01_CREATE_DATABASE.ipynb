```python
#create table 
def T_products():
    #read data
    products=pandas.read_csv("../NEW_DATASET/A01_TAB_PRODUCTS.csv",sep=';',decimal=',')
    
    #create table
    command="""CREATE TABLE PRODUCTS 
                (ID_PRODUCTS INTEGER UNIQUE,
                PROD_DESC VARCHAR,
                PROD_CLUSTER VARCHAR,
                UNITS_PACK INT,
                COST FLOAT,
                MARKUP FLOAT) """
    #cursor.execute(command) #create table in database

    #insert data
    command=''
    for register in products.values:
        register=f"""({register[0]},'{register[1]}','{register[2]}',{register[3]},{register[4]},{register[5]})"""
        command=command+','+register
    command=command[1:]
    command='INSERT INTO PRODUCTS VALUES '+command
    cursor.execute(command)
    banco.commit()

#T_products()
```

```python
def table_price():
    t_price=pandas.read_csv("../NEW_DATASET/A02_TAB_PRICE.csv",sep=';',decimal=',')
    
    #create table
    command="""CREATE TABLE TAB_PRICE 
                (TABLE_ID VARCHAR UNIQUE,
                ID_products INT,
                tab_name CHAR,
                unit_price FLOAT) """
    #cursor.execute(command)
    command=''
    for register in t_price.values:
        command=command+','+f"""('{register[0]}',{register[1]},'{register[2]}',{register[3]})"""
    command=command[1:]
    command="INSERT INTO TAB_PRICE VALUES "+command
    cursor.execute(command)
    banco.commit()
#table_price()
```
